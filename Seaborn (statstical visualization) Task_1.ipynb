```python
'''
---> Seaborn is a powerful tool used for statistical visulalization or distribution visualization
'''
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
vis1=sns.distplot(data.BirthRate)
# univariate analysis, used only one variable on x-axis
```

```python
vis1=sns.distplot(data.BirthRate,bins=4)
# --> The bins parameter enables you to control the bins of plot
# --> ex: if you set bins=20, the function will create a plot with 20 bars
```
